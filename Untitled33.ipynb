```python
from __future__ import annotations

import numpy as np
import pandas as pd
from tbats import TBATS


def load_series(csv_path: str, date_col: str = "date", value_col: str = "y") -> pd.Series:
    df = pd.read_csv(csv_path)
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values(date_col).set_index(date_col)

    # 日次の等間隔にそろえる
    # ここでは「土日=0」が仕様なので欠け日も0で埋める例
    s = df[value_col].asfreq("D").fillna(0.0)
    return s


def fit_forecast_tbats(
    y: np.ndarray,
    seasonal_periods: list[float],
    steps: int,
    n_jobs: int = 1,
    use_box_cox: bool = False,
    use_arma_errors: bool = True,
):
    estimator = TBATS(
        seasonal_periods=seasonal_periods,
        n_jobs=n_jobs,  # まずは1推奨（環境によっては並列で固まることがあります）
        use_box_cox=use_box_cox,
        use_arma_errors=use_arma_errors,
    )
    fitted = estimator.fit(y)
    fcst = fitted.forecast(steps=steps)
    return fcst, fitted
```

```python
# ===== 設定 =====
CSV_PATH = None  # 例: "/content/deposit_daily.csv"（Noneならダミーデータ）
DATE_COL = "date"
VALUE_COL = "y"

STEPS = 30               # 何日先まで
WEEKLY_ONLY = False      # True: [7] / False: [7, 30.5, 365.25]
N_JOBS = 1               # まずは1
USE_BOX_COX = False      # 0が入るならFalse推奨
USE_ARMA_ERRORS = True   # やや重いが精度が上がることがある
```

```python
if CSV_PATH:
    s = load_series(CSV_PATH, DATE_COL, VALUE_COL)
else:
    # ダミーデータ例：平日だけ動き、土日は0 + 少し季節性
    idx = pd.date_range("2024-01-01", periods=365, freq="D")
    base = 50 + 10 * np.sin(2 * np.pi * np.arange(len(idx)) / 30.5)
    s = pd.Series(base, index=idx)
    s[s.index.weekday >= 5] = 0.0  # 土日0
    s = s + np.random.normal(scale=3.0, size=len(s))

s.head(), s.tail()
```

```python
# --- 評価用の分割日（10月以降を比較）---
SPLIT_DATE = pd.Timestamp("2024-10-01")

# 念のため、日次に揃える（すでにasfreq("D")していれば不要ですが安全側）
s_daily = s.asfreq("D").fillna(0.0)

# train / test に分割
train = s_daily.loc[s_daily.index < SPLIT_DATE]
test  = s_daily.loc[s_daily.index >= SPLIT_DATE]

if len(test) == 0:
    raise ValueError("10月以降のデータがありません。SPLIT_DATE を調整してください。")

# 予測ステップ数は「テスト期間の長さ」
steps = len(test)

# 季節周期
seasonal_periods = [7] if WEEKLY_ONLY else [7, 30, 31]

# 学習（trainでfit）
y_train = train.values.astype(float)
fcst, model = fit_forecast_tbats(
    y=y_train,
    seasonal_periods=seasonal_periods,
    steps=steps,
    n_jobs=N_JOBS,
    use_box_cox=USE_BOX_COX,
    use_arma_errors=USE_ARMA_ERRORS,
)

# 予測のindex：train最終日の翌日からsteps日
fcst_index = pd.date_range(train.index[-1] + pd.Timedelta(days=1), periods=steps, freq="D")
pred = pd.Series(fcst, index=fcst_index, name="y_pred")

# 実績（test）と予測（pred）を同じ枠に揃える
cmp_df = pd.DataFrame({"y_true": test}).join(pred, how="left")

# 指標（参考）
mae = (cmp_df["y_true"] - cmp_df["y_pred"]).abs().mean()
rmse = np.sqrt(((cmp_df["y_true"] - cmp_df["y_pred"]) ** 2).mean())

mae, rmse, cmp_df.head()
```

```python
print(model.summary())

print("\n=== 2024/10/01 以降：実績 vs 予測（末尾） ===")
display(cmp_df.tail(20))

# グラフは 2024/06/01 以降だけ表示
PLOT_START = pd.Timestamp("2024-06-01")

plot_df = pd.DataFrame({"y": s_daily}).join(pred, how="left")
plot_df = plot_df.loc[plot_df.index >= PLOT_START]

import matplotlib.pyplot as plt
plt.figure()
plot_df["y"].plot(label="actual")
plot_df["y_pred"].plot(label="forecast (from Oct)")
plt.title("TBATS: actual vs forecast (plot from 2024-06-01)")
plt.legend()
plt.show()
```
